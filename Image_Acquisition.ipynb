ORANGE

```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread('orange.png')
cv2_imshow(img)
```

Gray Scale

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(img_gray)
```

Image Resize

```python
img_resize = cv2.resize(img_gray, (1000, 1000))
cv2_imshow(img_resize)

```

Image Flip

```python
img_flip = cv2.flip(img_gray, 0)
cv2_imshow(img_flip)
```
